##Data Preprocessing

```python
base_dir = "/content/drive/Shareddrives/capstone/dataset"
```

```python
##Preprocessing Image
IMAGE_SIZE = 150
BATCH_SIZE = 32

#Image Augmentation and Data Split
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255.,
    rotation_range=40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.1)

test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255.,
    validation_split=0.3)

train_datagen = train_datagen.flow_from_directory(
    base_dir, 
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode = 'categorical',
    subset='training')

test_datagen = test_datagen.flow_from_directory(
    base_dir, 
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode = 'categorical',
    subset='validation')
```

##Train Model

Get pre-trained model

```python
!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5 \
    -O /tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3

# Set the weights file you downloaded into a variable
local_weights_file = '/tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'

# Initialize the base model.
# Set the input shape and remove the dense layers.
pre_trained_model = InceptionV3(input_shape = (150, 150, 3), 
                                include_top = False, 
                                weights = None)

# Load the pre-trained weights you downloaded.
pre_trained_model.load_weights(local_weights_file)

# Freeze the weights of the layers.
for layer in pre_trained_model.layers:
  layer.trainable = False
```

```python
pre_trained_model.summary()
```

```python
# Choose `mixed_7` as the last layer of your base model
last_layer = pre_trained_model.get_layer('mixed7')
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

Create callback

```python
#Callback function when training accuracy reach 99.9%
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.999):
      print("\nReached 99.9% accuracy so cancelling training!")
      self.model.stop_training = True
```

Create final model

```python
#Model
#Consist of two dense layers with 512 and 256 neurons respectively, dropout layer, and output layer
x = layers.Flatten()(last_output)
x = tf.keras.layers.Dense(512, activation = 'relu')(x)
x = tf.keras.layers.Dense(256, activation = 'relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)  
x = tf.keras.layers.Dense(15, activation='softmax')(x)

model = Model(pre_trained_model.input, x) 

#Compiling model with Adam optimizer, Kullback Leibler Divergence Loss Function, and using accuracy metrics
model.compile(optimizer='adam',
              loss='kullback_leibler_divergence',
              metrics=['accuracy'])
    
```

```python
model.summary()
```

```python
#Model training and validation
#Using 40 epoch for optimal balance between accuracy, loss, and training time.
callbacks = myCallback()
history = model.fit(train_datagen, epochs=40, validation_data=test_datagen, callbacks=callbacks)
```

Plot accuracy

```python
#Plotting accuracy from training and validation to see if it is already good enough or there is anomaly.
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc=0)
plt.figure()

plt.show()
```

Plot Loss

```python
#Plotting loss function to see if the loss from training and validation normal or not.
fig = plt.figure()
plt.plot(history.history['loss'], color='teal', label='loss')
plt.plot(history.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

# Model Predict

```python
from google.colab import files
from tensorflow.keras.preprocessing import image
import numpy as np

# Step 1: Upload the image that want to be predicted
uploaded_files = files.upload()

# Step 2: List initialization to store predicted image
predictions = []

# Step 3: Looping sequence for predicting the image class
for image_path in uploaded_files.keys():
    # Image loading and preprocessing
    img = image.load_img(image_path, target_size=(150, 150, 3))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0

    # Predicting image class
    predictions = model.predict(img_array)

    # Dictionary for mapping class index into class label
    class_names = train_datagen.class_indices
    # Pulling class index with highest probability
    predicted_class_index = np.argmax(predictions)
    
    # Pulling class label related to predicted class index
    predicted_class = [k for k, v in class_names.items() if v == predicted_class_index][0]
    
    # Pulling the predicted class index probability
    predicted_prob = np.max(predictions)

    # Showing predicted image
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    
    # Showing prediction summary
    print('File:', image_path)
    print('Predicted class:', predicted_class)
    print('Probability:', predicted_prob)
    print()
```
